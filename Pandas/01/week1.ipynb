Importante: di default, DictReader consente di iteraresul file una sola volta.Ciò perchè DictReader legge il file sul momento senza immagazzinare in memoria.Ci sono dei modi per raggirare l'ostacolo ma non vengono trattati ora.La soluzione migliore è di caricare il file ogni volta che dobbiamo
usare dei valori presenti in esso

```python
import csv
with open('monumenti.csv', 'r') as csvfile:
    reader = csv.DictReader(csvfile)
    
    rows = 101
    for row in reader:
        rows -= 1
        if (rows > 0):
            print(row)
        else:
            break
```

Con le istruzioni sopra, si  effetta la lettura del file in un DictReader, eseguendo un'iterazione su di esso

```python
with open('monumenti.csv', 'r') as csvfile:
    reader = csv.DictReader(csvfile)
    
    monumenti_lista = []
    for row in reader:
        monumenti_lista.append(row)
        
    print("numero di monumenti: ", len(monumenti_lista))
```

Con quanto sopra, invece, si memorizza in una lista il contenuto del file, ogni riga del file è un elemento che si aggiunge alla lista.Possiamo determinare la lunghezza del file usando la funzione che ci restitisce la lunghezza della lista.Si nota che, facendo questo, non c'è più bisogno di riaprire il file e di rileggerne il suo contenuto in DictReader

```python
monumenti_2014 = [row for row in monumenti_lista if row["Annoinserimento"] == "2014"]
print("numero di monumenti posti nel 2014:", len(monumenti_2014))
```

Sopra è un esempio List Comprehensions: altro modo per aggiungere records alla lista.Metodo elegante e conciso.All'interno delle parentesi quadre, si intende " fa' qualcosa per ciascuna iterazione e usa il risultato, se vuoi, per aggiungerlo alla nuova lista".

```python
monumenti_2016 = [row for row in monumenti_lista if row["Annoinserimento"] == "2016"]
print("numero di monumenti posti nel 2016:", len(monumenti_2016))
for monumento in monumenti_2016:
    print ('A ',monumento["Comune"],"c'è " , monumento["Nome"])
```

Sopra è riportato un esempio: tra i dati, per un record, il campo "year" era scritto in lettere invece che in cifre.
Non trattandosi di dato numerico, non può essere convertito in intero e si genera un errore.Cerchiamo di prevenire una simile situazione con il codice sotto riportato.Prima andiamo a definire una funzione che controlla se un particolare valore può essere convertito ad un intero.Ciò viene effettuato con la cattura di un ValueError.


```python
def est_valido_anno(stringa):
    try:
        anno=int(stringa)
    except ValueError:
        return False
    else:
        return anno
```

```python
anno_di_inserimento=[int(row["Annoinserimento"]) for row in monumenti_lista if (est_valido_anno(row["Annoinserimento"]))]
```

```python
print (anno_di_inserimento)
```

```python
min_anno_inserimento=min(anno_di_inserimento)
print (min_anno_inserimento)
```
