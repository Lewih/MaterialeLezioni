<h1>Aggregate function </h1>

Per aggregare statistiche riassuntive diverse dalla media, si usa il parametro aggfunc per specificare le funzioni di aggregazione.Si usa, invece, il parametro values per specificare le colonne.Vogliamo sapere, nel nostro dataset, a quanto ammontano le recensioni, per ciascuna città.

```python
import numpy as np
```

```python
pivot_agg=pd.pivot_table(
    df,index=["state","city"],
     values=["review_count"],
     aggfunc=[np.sum]
)
```

```python
pivot_agg
```

Sopra abbiamo risposto alla domanda, quale sommadi recensioni per ogni città, raggruppando le città stesse per stato.Ora, invece, di seguito, vogliamo distinguere, nel calcolare la somma recensioni, quante siano per gli esercizi con take_out e quante qelle per esercizi che non effettuano tale esercizi

```python
pivot_a2=pd.pivot_table(
    df,index=["state","city"],
    values=["review_count"],
    columns=["take_out"], #specifichiamo le colonne che separano i risultati
    aggfunc=[np.sum]
    )
```

```python
pivot_a2
```

Possiamo vedere il numero di recensioni per città fornito in modo separato in base a take_out.Nell'esempio sotto, invece, il parametro aggfunc è un dizionario comodo per indicare su quali colonne agire con quale funzione (somma o media)

```python
pivot_agg3=pd.pivot_table(
    df,index=["state","city"],
    columns=["take_out"],
    aggfunc={"review_count":np.sum,"stars":np.mean})
```

```python
pivot_agg3
```
